```python
import matplotlib.pyplot as plt
import numpy as np

# t 값 생성 (하트 곡선의 매개변수)
t = np.linspace(0, 2 * np.pi, 1000)

# 하트 방정식 (파라메트릭)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 그래프 그리기
plt.figure(figsize=(6,6))
plt.plot(x, y, color='red')
plt.title("❤️ 하트 모양 그래프", fontsize=16)
plt.axis('off')  # 축 숨기기
plt.show()
```
